#### Posición

Las medidas de posición estiman qué valores dividen la población en unas determinadas proporciones. Los valores que determinan estas posiciones reciben el nombre de cuantiles. En este sentido, la mediana es también una medida de posición, puesto que divide la variable en dos mitades, que sería el cuantil/cuartil 0.5.

Se diferencian los siguientes cuantiles:

+ La **mediana** es el cuantil $Q_0.5$.
+ Los **cuartiles** son los cuantiles $Q_{0.25}$, $Q_{0.5}$ y $Q_{0.75}$, y reciben, respectivamente, los nombres de primer cuartil, segundo cuartil (o mediana) y tercer cuartil. $Q_{0.25}$ será, pues, el menor valor que es mayor o igual que una cuarta parte de los datos, y $Q_{0.75}$, el menor valor que es mayor o igual que tres cuartas partes de los datos.
+ Los **deciles** son los cuantiles $Q_p$ con $p$ un múltiplo entero de 0.1: el primer decil es $Q_{0.1}$, el segundo decil es $Q_{0.2}$, y así sucesivamente.
+ Los **percentiles** son los cuantiles $Q_p$ con $p$ un múltiplo entero de 0.01.

#### Dispersión

Las medidas de dispersión evalúan lo disperso que están los datos. Las más importantes son:

+ El **rango**, o recorrido, que es la diferencia entre el máximo y el mínimo de las observaciones


+ El **rango intercuartílico**, que es la diferencia $Q_{0.75}−Q_{0.25}$. También se llama a veces rango intercuartílico al intervalo intercuartílico.


+ La **varianza**, que es la media aritmética de las diferencias al cuadrado entre los datos $x_i$ y la media $\bar x$
de la variable y la denotamos por $s^2$.


+ La **desviación típica**, que es la raíz cuadrada positiva $s$ de la varianza: $s=\sqrt s^2$. 

Las medidas de dispersión de la columna $numero$_$bateos$ son:

#### Forma

Las medidas de forma miden la mayor o menor deformación de la distribución con respecto a la distribución normal. Existen dos medidas de forma fundamentales:

1. El **coeficiente de asimetría** es un indicador del grado de igualdad de las "dos mitades" de la distribución. La interpretación de los valores que obtenemos del coeficiente de asimetría **$g_1$ de Fisher (skewness)** sería:
   - g_1 = 0 --> Distribución simétrica
   - g_1 > 0 --> Distribución asimétrica positiva o a la derecha
   - g_1 < 0 --> Distribución asimétrica negativa o a la izquierda

2. El **coeficiente de curtosis** mide el grado de apuntamiento o achatamiento de la distribución respecto a la normal. La interpretación de los valores que obtenemos del coeficiente de curtosis **$g_2$ (kurtosis)** sería:
   - g_2 = 0 --> Distribución mesocúrtica
   - g_2 < 0 --> Distribución platicúrtica
   - g_2 > 0 --> Distribución leptocúrtica

En las siguientes celdas se van a obtener ambos coeficientes, para su posterior interpretación:

```r
library(e1071)
```

Coeficiente de asimetría y de curtosis de la columna $numero$_$bateos$:

```r
skewness(beisbol$numero_bateos)
kurtosis(beisbol$numero_bateos)
```

Lo que se puede interpretar de cada coeficiente es:

- Coeficiente de asimetría: Al ser el resultado 0.46229, g_1 > 0, por tanto la columna $numero$_$bateos$ es asimétrica positiva o a la derecha, señalando que la mayoría de datos se concentran en la parte izquierda de la distribución, indicando que existen valores atípicos o extremadamente altos en esta zona (algunos equipos tienen un rendimiento notablemente mayor a otros).
  
- Coeficiente de curtosis: Al ser el resultado -0.6447, g_2 < 0, por tanto la columna $numero$_$bateos$ tiene una distribución platicúrtica. Esta distribución es la que señala mayor desigualdad entre datos. Cuanto más se acerque a -1 el coeficiente, mayor será la desigualdad, ya que los datos no se distribuyen alrededor de la media, como pasa en la distribución mesocúrtica y leptocúrtica. En este caso se puede confirmar que existen datos notablemente desiguales entre sí, sin ser extremadamente diferentes.

Coeficiente de asimetría y de curtosis de la columna $runs$:

```r
skewness(beisbol$runs)
kurtosis(beisbol$runs)
```

De estos coeficientes se llega a las mismas conclusiones que en el apartado anterior, ya que el coeficiente de simetría es positivo, y el de curtosis es negativo, por tanto la columna $runs$ es asimétrica positiva y de distribución platicúrtica.

Es importante aclarar que en el coeficiente de asimetría es mayor que el del caso anterior, por tanto se puede afirmar que hay mayor cantidad de datos agrupados en la derecha de la distribución. En cambio, tiene un valor menor de coeficiente de curtosis, indicando que siguen existiendo desigualdad entre datos, pero más leve, tendiendo los datos más a la media.

Para comprobar que existe una linealidad entre las variables que se quieren trabajar, se va a comenzar por utilizar el método $scatter.smooth$, que devolverá un gráfico donde se puede observar si los datos se relacionan entre sí o no:

Tras interpretar el gráfico, se puede confirmar que no se trata de un patrón aleatorio (ya que los puntos se distribuyen alrededor de la línea, y no están dispersados aleatoriamente). Esto es una evidencia que puede existir una linealidad entre las variables.

Para poder confirmar dicha linealidad, se va a utilizar la función $cor$, que devolverá el coeficiente de correlación entre las variables introducirdas, en este caso, las columnas de $runs$ y $numero$_$bateos$.

```r
cor(beisbol$runs, beisbol$numero_bateos)
```

El coeficiente de correlación es distinto a cero y positivo, señalando que si existe una relación entre las variables, y además que tienen una relación directa entre estas, es decir, si hay un mayor número de bateos, habrá un mayor número de runs y viceversa.

Para poder confirmar esto, es necesario realizar un contraste de correlación, utilizando el método $cor.test$:

```r
cor.test(beisbol$runs, beisbol$numero_bateos)
```

Como se puede observar, el $p-valor$ obtenido es igual a 0.00033, que es menor que el nivel de significación $\alpha = 0.05$, por tanto, se rechazará la hipótesis nula h_0, la cuál señalaba que el coeficiente de correlación entre estas variables es igual a cero. Al rechazarla se confirma que las variables SI están correlacionadas, en este caso, como se ha comentado anteriormente, tienen una relación directa. 

Esto permite realizar la estimación de un modelo de regresión lineal entre dichas variables.

<div style="background-color:#5DADE2"><b>Apartado 3</b></div>

Para estimar el modelo de regresión lineal simple entre la variable dependiente $runs$ y la independiente $numero$_$bateos$ se va a utilizar el método $lm$ como se puede observar en el siguiente input:

```r
# Alternativa 1 incluyendo el dataset como parámetro
modelo <- lm(runs ~ numero_bateos, data = beisbol)  
modelo
```

Los coeficientes obtenidos son b_0 = -2789.2429, y b_1 = 0.6305. B_0 representa el valor de la intersección con el eje y, indicando el número de runs cuando el número de bateos es cero, es decir, el coeficiente del número de runs base sin tener en cuenta el número de bateos.
b_1 es el coeficiente de la pendiente, y muestra cómo cambia el número de runs en función del número de bateos. 
En este caso, por cada unidad adicional de número de bateos, se espera un aumento de 0.6305 unidades de runs. Al ser este un coeficiente positivo, se confirma que hay una relación directa y proporcional entre las variables como se ha visto en apartados anteriores, indicando que un número mayor de runs se asocia a un número mayor de bateos.

El coeficiente de determinación $R^2$, es otra de las medidas fundamentales en la interpretación de un modelo de regresión lineal simple. 

En este caso, donde la variable dependiente es el número de runs y la variable independiente es el número de bateos, un $R^2$ igual a $0.3728$ significa que aproximadamente el 37.28% de la **variabilidad en el número de runs puede ser explicada por el número de bateos**. 

Este valor implica que, aunque existe una relación entre número de bateos y número de runs, hay otros factores significativos que también influyen en el número de runs y que no están incluidos en el modelo. 

El $R^2$ de este tamaño sugiere que, aunque el número de bateos es un factor importante, es solo uno de varios elementos que determinan el número de runs por equipo, dejando un porcentaje de 67.72% de la variabilidad en los números de runs por equipo atribuible a otros factores no contemplados en el modelo.

Hay que tener en cuenta que si se añaden más variables independientes al modelo, este mejorará por el hecho de incluir más variables al modelo. Por ello $R^2$ tiende a sobreestimar el proder predictivo del modelo, aunque las variables no sean significativas.

Al calcular el coeficiente de determinación ajustado $R^2_{adj}$ se aplica una penalización a la incorporación de nuevas variables independientes:

```r
# Obtenemos R2 ajustado directamente del output (NO del modelo)
R2adj=output$adj.r.squared
R2adj
```

```r
# Obtenemos R2 ajustado a través de su ecuación formal 1-(1-R2)*(n-1/n-k)
# k = número parámetros estimados
# n = número de observaciones
R2adj_2=1-(1-R2)*((526-1)/(526-2))
R2adj_2
```

Esto da un número sin sobreestimar el poder predictivo del modelo, aunque como se puede observar, los coeficientes son muy parejos, por lo que en este caso no sería de gran relevancia.

#### Forma

Las medidas de forma miden la mayor o menor deformación de la distribución con respecto a la distribución normal. Existen dos medidas de forma fundamentales:

1. El **coeficiente de asimetría** es un indicador del grado de igualdad de las "dos mitades" de la distribución. La interpretación de los valores que obtenemos del coeficiente de asimetría **$g_1$ de Fisher (skewness)** sería:
   - g_1 = 0 --> Distribución simétrica
   - g_1 > 0 --> Distribución asimétrica positiva o a la derecha
   - g_1 < 0 --> Distribución asimétrica negativa o a la izquierda

2. El **coeficiente de curtosis** mide el grado de apuntamiento o achatamiento de la distribución respecto a la normal. La interpretación de los valores que obtenemos del coeficiente de curtosis **$g_2$ (kurtosis)** sería:
   - g_2 = 0 --> Distribución mesocúrtica
   - g_2 < 0 --> Distribución platicúrtica
   - g_2 > 0 --> Distribución leptocúrtica

En las siguientes celdas se van a obtener ambos coeficientes, para su posterior interpretación:

```r
library(e1071)
```

```r
skew_analisis = function(coef){
    if(coef == 0){
        return(cat("Análisis del coeficiente de asimetría: Los datos tienen una distribución simétrica"))
    } else if(coef > 0) {
        return(cat("Análisis del coeficiente de asimetría: Los datos tienen una distribución positiva o a la derecha"))
    } else {
        return(cat("Análisis del coeficiente de asimetría: Los datos tienen una distribución negativa o a la izquierda"))
    }
}

kurt_analisis = function(coef){
    if(coef == 0){
        return(cat("Análisis del coeficiente de curtosis: Los datos tienen una distribución mesocúrtica"))
    } else if (coef > 0) {
        return(cat("Análisis del coeficiente de curtosis: Los datos tienen una distribución leptocúrtica"))
    } else {
        return(cat("Análisis del coeficiente de curtosis: Los datos tienen una distribución platicúrtica"))
    }
}
```

```r
for (col in colnames(datos_usa[-1])) {
    cat("Medidas de forma de la columna:", col,"\n")
    
    cat("Coeficiente de asimetría:", skewness(datos_usa[[col]]),"\n")
    cat(skew_analisis(skewness(datos_usa[[col]])),"\n")
    
    cat("Coeficiente de curtosis:", kurtosis((datos_usa[[col]])),"\n")
    cat(kurt_analisis(kurtosis(datos_usa[[col]])),"\n")
    cat("-------------------------------","\n")
}
```

```r
# Criterios de selección del modelo en base al ajuste de los datos
```

Para poder crear el mejor modelo que consiga ajustar sus predicciones lo máximo posible a la variable dependiente $obama$, hay que comparar modelos cambiando las variables independientes que se disponen dentro de los datos del dataframe original. Los criterios de selección de modelos con los que se decidirá que modelo es el mejor son:

1. Comparación del **coeficiente de determinación ajustado**: el modelo con un mayor $R^2_{adj}$ será el mejor de los comparados. 


2. Comparación de los **criterios de información** de Akaike (AIC) y Bayesiano (BIC): cuanto menor sea el valor, mejor será el ajuste realizado. La diferencia entre AIC y BIC es la severidad con la que penalizan el número de parámetros considerados. La fórmula de cada uno de ellos es la siguiente:

- $AIC = n\ln \left( {SSR/n} \right) + 2k$

- $BIC = n\ln \left( {SSR/n} \right) + k\ln \left( n \right)$

Al tratarse de un caso múltiple, lo más lógico es pasar directamente a la automatización de la búsqueda del mejor modelo, pero se van a realizar distintos modelos manualmente para poder hacerse una idea de las variables independientes que pueden terminar siendo las más importantes.

```r
names(datos_usa)
```

```r
datos_usa2 = datos_usa[-1] # Se elimina la columna de estado ya que no se va a utilizar en la selección del mejor modelo
```

```r
# Creación de modelos 
mod1=lm(obama ~ grad_univ+afro+blancos, data = datos_usa2)
mod2=lm(obama ~ grad_univ*afro*blancos, data = datos_usa2)
mod3=lm(obama ~ salario+jubilados+veteranos, data = datos_usa2)
mod4=lm(obama ~ salario*jubilados*veteranos, data = datos_usa2)
mod5=lm(obama ~ evangelicos+protestantes+catolicos, data = datos_usa2)
mod6=lm(obama ~ evangelicos*protestantes*catolicos, data = datos_usa2)
mod7=lm(obama ~ afro+asia+hispanos+blancos, data = datos_usa2)
mod8=lm(obama ~ afro*asia*hispanos*blancos, data = datos_usa2)
mod9=lm(obama ~ ., data = datos_usa2) # tenemos en cuenta todas las variables independientes
```

```r
# Calculamos los 3 criterios de selección de modelos

# Coeficiente de determinación ajustado
R2=c(summary(mod1)$adj.r.squared, summary(mod2)$adj.r.squared,summary(mod3)$adj.r.squared,
     summary(mod4)$adj.r.squared,summary(mod5)$adj.r.squared,summary(mod6)$adj.r.squared,
     summary(mod7)$adj.r.squared,summary(mod8)$adj.r.squared,summary(mod9)$adj.r.squared)

# Criterio de Información de Akaike
AIC=c(extractAIC(mod1)[2],extractAIC(mod2)[2],extractAIC(mod3)[2],extractAIC(mod4)[2],
      extractAIC(mod5)[2],extractAIC(mod6)[2],extractAIC(mod7)[2],extractAIC(mod8)[2],
      extractAIC(mod9)[2])

# Criterio de Información Bayesiano
BIC=c(extractAIC(mod1,k=log(nrow(datos_usa2)))[2],extractAIC(mod2,k=log(nrow(datos_usa2)))[2],
      extractAIC(mod3,k=log(nrow(datos_usa2)))[2],extractAIC(mod4,k=log(nrow(datos_usa2)))[2],
      extractAIC(mod5,k=log(nrow(datos_usa2)))[2],extractAIC(mod6,k=log(nrow(datos_usa2)))[2],
      extractAIC(mod7,k=log(nrow(datos_usa2)))[2],extractAIC(mod8,k=log(nrow(datos_usa2)))[2],
      extractAIC(mod9,k=log(nrow(datos_usa2)))[2])
```

```r
# Organizaremos los valores de los diferentes criterios en un dataframe para facilitar su comparación
# Usaremos la función `kable` de la libería knitr para obtener una tabla bien formateada
require(knitr)
```

Para poder confirmar que modelo es el mejor, hay que encontrar el mayor coeficiente de determinación ajustado $R^2_{adj}$, y el menor AIC y BIC.

Significa pues que el modelo9 es el que más se ajusta a $obama$, modelo que tiene en cuenta todas las variables de la tabla. A pesar de ello, no todas las variables son significativas, por lo que ahora habrá que proceder a refinar el modelo de la siguiente forma.

```r
# Proceso de automatización del proceso de selección del mejor modelo
```

Para poder automatizar el proceso de selección del mejor modelo se puede utilizar el método $step$, el cual a partir de un modelo lineal dato prueba diferentes modelos añadiendo o eliminando variables independientes hasta encontrar el modelo óptimo.

Los otros parámetros son $direction$, que indica la metodología para generar nuevos modelos, $scope$, que define el rando de modelos que se tienen que considerar, y $k$, que indica si la evaluación de los modelos se realiza en base al AIC o al BIC.

```r
# 1º Tenemos que estimar el modelo con todas las variables
mod9=lm(obama ~ ., data = datos_usa2)
summary(mod9)
```

```r
# 2º Búsqueda automática del modelo óptimo hacia atrás
step(mod9,direction="backward")
```

El algoritmo comienza con el modelo inicial que contiene todas las variables. Por cada paso que se realiza, se comprueba cual es la mejor variable independiente a eliminar para conseguir un mejor AIC. Así de manera progresiva el algoritmo elimina variables hasta que $none$ es la que menor AIC tiene. $none$ indica que no se puede conseguir ningún AIC menor eliminando variables, y por tanto se ha encontrado el modelo óptimo.

En este caso el modelo óptimo contiene las variables independientes $densidad$, $mujeres$, $grad$_ $instituto$, $grad$_ $univ$, $asia$, $hispanos$, $evangelicos$, $mormones$ y $paro$

Ahora se va a probar con la opción $forward$, el cuál irá añadiendo variables independientes hasta que no se pueda conseguir un AIC menor.

Para ello, primero hay que crear un modelo sin variables independientes, tal y como se ve en la siguiente celda:

```r
mod_cte= lm(obama~1,data=datos_usa2)
summary(mod_cte)
```

```r
# Se ejecuta step en modo forward
step(mod_cte,direction="forward",scope=list(lower=mod_cte, upper=mod9))
```

Como resultado, el método forward indica el modelo óptimo contiene las variables independientes $grad_univ$, $paro$, $jubilados$, $asia$, $mormones$, $densidad$, $evangelicos$, $mujeres$, y $protestantes$.

Esto indica que los modelos óptimos no coinciden totalmente, ya que por ejemplo $jubilados$ y $protestantes$ no están en el modelo de backward y si de forward.

Por tanto, para decidir que modelo es mejor, habrá que comparar el AIC final de cada uno. En el caso del primero, el AIC es de 165.22, del segundo es de 166.02.

Entonces, el módelo óptimo es el conseguido a través del método $backward$.

5 - Ahora se extraen todos los datos que hay que devolver al final de la función, como por ejemplo: 

- Valores ajustados: Se toma como ejemplo la alternativa 1 del boletín 3.3 del cáculo de los valores ajustados manual, y se ha realizado un bucle for que recorre las variables independientes con el fin de calcular todos los valores ajustados.
- Grados de libertad: Se calculan restando el número de observaciones menos el número de variables independeientes menos 1.
- Residuos: Se obtienen restando los valores originales de $Girth$ por los valores predichos por el modelo.
- Suma de cuadrados de los residuos: Raíz cuadrada del sumatorio de los residuos al cuadrado entre los grados de libertad.

6 - Se agrupan todos los datos obtenidos en una lista, para que el usuario del método pueda utilizar estos datos de manera fácil.

7 - Se crea un output, donde se puede dar un vistazo rápido de todos los datos obtenidos, como se consigue con el método $summary$ de R

```r
formula = Girth ~ Height + Volume
variables = all.vars(as.formula(formula))
variables[1]
```

```r
lm_ej5 <- function(data, formula, alpha = 0.05) {
    
    # 0. Declaración de la función y sus parámetros
    # data:  dataframe que contiene los datos
    # formula: fórmula del modelo de regresión lineal
    # alpha: nivel de significación para los intervalos de confianza 
  
    # 1. Extraer los nombres de las variables de la fórmula con el metodo all.vars
    variables = all.vars(as.formula(formula))
    variable_dependiente = variables[1]
    variables_independientes = variables[-1]
    
    # 2. Creación de la matriz X con las variables independientes con B_0s, y el vector y con la variable dependiente
    X = as.matrix(data[, variables[-1]])  # Excluye la variable dependiente
    # Agregar una columna de unos para los B_0
    X = cbind(1, X)
    # Vector y igual a la columna Girth original
    y = as.matrix(data[[variables[1]]])
  
    # 3. Calcular los coeficientes usando la fórmula matricial: β = (X^T * X)^(-1) * X^T * y
    coeficientes = solve(t(X) %*% X) %*% (t(X) %*% y)
  
    # 4. Número de observaciones y número de variables independientes (predictores)
    n = nrow(X)
    p = length(variables) - 1 

    # 5. Extraccion de datos

    # Valores ajustados
    val_ajustado = rep(coeficientes[1], n)
    # Basado en ejemplo boletin 3.3 yajustado <- bhat[1]+bhat[2]*wdata$educ+bhat[3]*wdata$exper+bhat[4]*wdata$tenure
    for (i in seq_along(variables_independientes)) {
        val_ajustado <- val_ajustado + coeficientes[i + 1] * data[[variables_independientes[i]]]
    }
    
    # Grados de libertad
    gdl = n - p - 1
  
    # Residuos
    residuos = y - X %*% coeficientes
  
    # Suma de cuadrados de los residuos
    ssr = sqrt(sum(residuos^2)/gdl)

    # Coeficientes de determinacion, No funciona porque al hacer la media aparece NA por algun motivo
    # residuos2 = data[variable_dependiente] - mean(data[variable_dependiente])
    # coef_det = 1-(sum(residuos^2)/sum((residuos2)^2))
          
    # 6. Crear un objeto de modelo para devolver
    modelo = list(
        coefficients = coeficientes,
        residuals = residuos,
        gdl = gdl,
        ssr = ssr,
        val_ajustado = val_ajustado
        #coef_det = coef_det
    )

    # 7. Output
    cat("Call:", "\n", formula, "\n")
    cat("Residuals:", "\n", "Min:", min(residuos), "1Q:", quantile(residuos, 0.25),
        "Median:", median(residuos), "3Q:", quantile(residuos, 0.75), "Max:", max(residuos), "\n")
    cat("Coefficients:", "\n", coeficientes, "\n")
    cat("Degrees of Freedom:", gdl, "\n")
    cat("Residual Standar Error:", ssr, "\n")
    #cat("Multiple R-squared:", coef_det, "\n")

    
    # Devolver el modelo ajustado
    return(modelo)
}

# Ejemplo de uso:
form = "Girth ~ Height + Volume"  # Ajusta esto según tus nombres de variables
mod_ej5_func = lm_ej5(data_ej5, form)
cat("Los valores ajustados son:","\n",mod_ej5_func$val_ajustado)
```
