#### Posicin

Las medidas de posicin estiman qu valores dividen la poblacin en unas determinadas proporciones. Los valores que determinan estas posiciones reciben el nombre de cuantiles. En este sentido, la mediana es tambin una medida de posicin, puesto que divide la variable en dos mitades, que sera el cuantil/cuartil 0.5.

Se diferencian los siguientes cuantiles:

+ La **mediana** es el cuantil $Q_0.5$.
+ Los **cuartiles** son los cuantiles $Q_{0.25}$, $Q_{0.5}$ y $Q_{0.75}$, y reciben, respectivamente, los nombres de primer cuartil, segundo cuartil (o mediana) y tercer cuartil. $Q_{0.25}$ ser, pues, el menor valor que es mayor o igual que una cuarta parte de los datos, y $Q_{0.75}$, el menor valor que es mayor o igual que tres cuartas partes de los datos.
+ Los **deciles** son los cuantiles $Q_p$ con $p$ un mltiplo entero de 0.1: el primer decil es $Q_{0.1}$, el segundo decil es $Q_{0.2}$, y as sucesivamente.
+ Los **percentiles** son los cuantiles $Q_p$ con $p$ un mltiplo entero de 0.01.

Las medidas de dispersin evalan lo disperso que estn los datos. Las ms importantes son:

+ La **varianza**, que es la media aritmtica de las diferencias al cuadrado entre los datos $x_i$ y la media $\bar x$

+ La **desviacin tpica**, que es la raz cuadrada positiva $s$ de la varianza: $s=\sqrt s^2$.

#### Forma

Las medidas de forma miden la mayor o menor deformacin de la distribucin con respecto a la distribucin normal. Existen dos medidas de forma fundamentales:

1. El **coeficiente de asimetra** es un indicador del grado de igualdad de las "dos mitades" de la distribucin. La interpretacin de los valores que obtenemos del coeficiente de asimetra **$g_1$ de Fisher (skewness)** sera:
   - g_1 = 0 --> Distribucin simtrica
   - g_1 > 0 --> Distribucin asimtrica positiva o a la derecha
   - g_1 < 0 --> Distribucin asimtrica negativa o a la izquierda

2. El **coeficiente de curtosis** mide el grado de apuntamiento o achatamiento de la distribucin respecto a la normal. La interpretacin de los valores que obtenemos del coeficiente de curtosis **$g_2$ (kurtosis)** sera:
   - g_2 = 0 --> Distribucin mesocrtica
   - g_2 < 0 --> Distribucin platicrtica
   - g_2 > 0 --> Distribucin leptocrtica

En las siguientes celdas se van a obtener ambos coeficientes, para su posterior interpretacin:

```r
library(e1071)
```

Coeficiente de asimetra y de curtosis de la columna $numero$_$bateos$:

```r
skewness(beisbol$numero_bateos)
kurtosis(beisbol$numero_bateos)
```

Lo que se puede interpretar de cada coeficiente es:

- Coeficiente de asimetra: Al ser el resultado 0.46229, g_1 > 0, por tanto la columna $numero$_$bateos$ es asimtrica positiva o a la derecha, sealando que la mayora de datos se concentran en la parte izquierda de la distribucin, indicando que existen valores atpicos o extremadamente altos en esta zona (algunos equipos tienen un rendimiento notablemente mayor a otros).
  
- Coeficiente de curtosis: Al ser el resultado -0.6447, g_2 < 0, por tanto la columna $numero$_$bateos$ tiene una distribucin platicrtica. Esta distribucin es la que seala mayor desigualdad entre datos. Cuanto ms se acerque a -1 el coeficiente, mayor ser la desigualdad, ya que los datos no se distribuyen alrededor de la media, como pasa en la distribucin mesocrtica y leptocrtica. En este caso se puede confirmar que existen datos notablemente desiguales entre s, sin ser extremadamente diferentes.

Coeficiente de asimetra y de curtosis de la columna $runs$:

```r
skewness(beisbol$runs)
kurtosis(beisbol$runs)
```

De estos coeficientes se llega a las mismas conclusiones que en el apartado anterior, ya que el coeficiente de simetra es positivo, y el de curtosis es negativo, por tanto la columna $runs$ es asimtrica positiva y de distribucin platicrtica.

Es importante aclarar que en el coeficiente de asimetra es mayor que el del caso anterior, por tanto se puede afirmar que hay mayor cantidad de datos agrupados en la derecha de la distribucin. En cambio, tiene un valor menor de coeficiente de curtosis, indicando que siguen existiendo desigualdad entre datos, pero ms leve, tendiendo los datos ms a la media.

#### Forma

Las medidas de forma miden la mayor o menor deformacin de la distribucin con respecto a la distribucin normal. Existen dos medidas de forma fundamentales:

1. El **coeficiente de asimetra** es un indicador del grado de igualdad de las "dos mitades" de la distribucin. La interpretacin de los valores que obtenemos del coeficiente de asimetra **$g_1$ de Fisher (skewness)** sera:
   - g_1 = 0 --> Distribucin simtrica
   - g_1 > 0 --> Distribucin asimtrica positiva o a la derecha
   - g_1 < 0 --> Distribucin asimtrica negativa o a la izquierda

2. El **coeficiente de curtosis** mide el grado de apuntamiento o achatamiento de la distribucin respecto a la normal. La interpretacin de los valores que obtenemos del coeficiente de curtosis **$g_2$ (kurtosis)** sera:
   - g_2 = 0 --> Distribucin mesocrtica
   - g_2 < 0 --> Distribucin platicrtica
   - g_2 > 0 --> Distribucin leptocrtica

En las siguientes celdas se van a obtener ambos coeficientes, para su posterior interpretacin:

```r
library(e1071)
```

```r
skew_analisis = function(coef){
    if(coef == 0){
        return(cat("Anlisis del coeficiente de asimetra: Los datos tienen una distribucin simtrica"))
    } else if(coef > 0) {
        return(cat("Anlisis del coeficiente de asimetra: Los datos tienen una distribucin positiva o a la derecha"))
    } else {
        return(cat("Anlisis del coeficiente de asimetra: Los datos tienen una distribucin negativa o a la izquierda"))
    }
}

kurt_analisis = function(coef){
    if(coef == 0){
        return(cat("Anlisis del coeficiente de curtosis: Los datos tienen una distribucin mesocrtica"))
    } else if (coef > 0) {
        return(cat("Anlisis del coeficiente de curtosis: Los datos tienen una distribucin leptocrtica"))
    } else {
        return(cat("Anlisis del coeficiente de curtosis: Los datos tienen una distribucin platicrtica"))
    }
}
```

```r
for (col in colnames(datos_usa[-1])) {
    cat("Medidas de forma de la columna:", col,"\n")
    
    cat("Coeficiente de asimetra:", skewness(datos_usa[[col]]),"\n")
    cat(skew_analisis(skewness(datos_usa[[col]])),"\n")
    
    cat("Coeficiente de curtosis:", kurtosis((datos_usa[[col]])),"\n")
    cat(kurt_analisis(kurtosis(datos_usa[[col]])),"\n")
    cat("-------------------------------","\n")
}
```

Los criterios de seleccin de modelos con los que se decidir que modelo es el mejor son:

1. Comparacin del **coeficiente de determinacin ajustado**: el modelo con un mayor $R^2_{adj}$ ser el mejor de los comparados. 


2. Comparacin de los **criterios de informacion** de Akaike (AIC) y Bayesiano (BIC): cuanto menor sea el valor, mejor ser el ajuste realizado. La diferencia entre AIC y BIC es la severidad con la que penalizan el numero de parametros considerados. La frmula de cada uno de ellos es la siguiente:

- $AIC = n\ln \left( {SSR/n} \right) + 2k$

- $BIC = n\ln \left( {SSR/n} \right) + k\ln \left( n \right)$

Al tratarse de un caso mltiple, lo ms lgico es pasar directamente a la automatizacin de la bsqueda del mejor modelo, pero se van a realizar distintos modelos manualmente para poder hacerse una idea de las variables independientes que pueden terminar siendo las ms importantes.

```r
# Creacin de modelos 
mod1=lm(obama ~ grad_univ+afro+blancos, data = datos_usa2)
mod2=lm(obama ~ grad_univ*afro*blancos, data = datos_usa2)
mod3=lm(obama ~ salario+jubilados+veteranos, data = datos_usa2)
mod4=lm(obama ~ salario*jubilados*veteranos, data = datos_usa2)
mod5=lm(obama ~ evangelicos+protestantes+catolicos, data = datos_usa2)
mod6=lm(obama ~ evangelicos*protestantes*catolicos, data = datos_usa2)
mod7=lm(obama ~ afro+asia+hispanos+blancos, data = datos_usa2)
mod8=lm(obama ~ afro*asia*hispanos*blancos, data = datos_usa2)
mod9=lm(obama ~ ., data = datos_usa2) # tenemos en cuenta todas las variables independientes
```

```r
# Calculamos los 3 criterios de seleccin de modelos

# Coeficiente de determinacin ajustado
R2=c(summary(mod1)$adj.r.squared, summary(mod2)$adj.r.squared,summary(mod3)$adj.r.squared,
     summary(mod4)$adj.r.squared,summary(mod5)$adj.r.squared,summary(mod6)$adj.r.squared,
     summary(mod7)$adj.r.squared,summary(mod8)$adj.r.squared,summary(mod9)$adj.r.squared)

# Criterio de Informacin de Akaike
AIC=c(extractAIC(mod1)[2],extractAIC(mod2)[2],extractAIC(mod3)[2],extractAIC(mod4)[2],
      extractAIC(mod5)[2],extractAIC(mod6)[2],extractAIC(mod7)[2],extractAIC(mod8)[2],
      extractAIC(mod9)[2])

# Criterio de Informacin Bayesiano
BIC=c(extractAIC(mod1,k=log(nrow(datos_usa2)))[2],extractAIC(mod2,k=log(nrow(datos_usa2)))[2],
      extractAIC(mod3,k=log(nrow(datos_usa2)))[2],extractAIC(mod4,k=log(nrow(datos_usa2)))[2],
      extractAIC(mod5,k=log(nrow(datos_usa2)))[2],extractAIC(mod6,k=log(nrow(datos_usa2)))[2],
      extractAIC(mod7,k=log(nrow(datos_usa2)))[2],extractAIC(mod8,k=log(nrow(datos_usa2)))[2],
      extractAIC(mod9,k=log(nrow(datos_usa2)))[2])
```

```r
# Proceso de automatizacin del proceso de seleccin del mejor modelo
```

Para poder automatizar el proceso de seleccin del mejor modelo se puede utilizar el mtodo $step$, el cual a partir de un modelo lineal dato prueba diferentes modelos aadiendo o eliminando variables independientes hasta encontrar el modelo ptimo.

Los otros parmetros son $direction$, que indica la metodologa para generar nuevos modelos, $scope$, que define el rando de modelos que se tienen que considerar, y $k$, que indica si la evaluacin de los modelos se realiza en base al AIC o al BIC.

```r
# 1 Tenemos que estimar el modelo con todas las variables
mod9=lm(obama ~ ., data = datos_usa2)
summary(mod9)
```

```r
# 2 Bsqueda automtica del modelo ptimo hacia atrs
step(mod9,direction="backward")
```

El algoritmo comienza con el modelo inicial que contiene todas las variables. Por cada paso que se realiza, se comprueba cual es la mejor variable independiente a eliminar para conseguir un mejor AIC. As de manera progresiva el algoritmo elimina variables hasta que $none$ es la que menor AIC tiene. $none$ indica que no se puede conseguir ningn AIC menor eliminando variables, y por tanto se ha encontrado el modelo ptimo.

En este caso el modelo ptimo contiene las variables independientes $densidad$, $mujeres$, $grad$_ $instituto$, $grad$_ $univ$, $asia$, $hispanos$, $evangelicos$, $mormones$ y $paro$

Ahora se va a probar con la opcin $forward$, el cul ir aadiendo variables independientes hasta que no se pueda conseguir un AIC menor.

Para ello, primero hay que crear un modelo sin variables independientes, tal y como se ve en la siguiente celda:

```r
mod_cte= lm(obama~1,data=datos_usa2)
summary(mod_cte)
```

```r
# Se ejecuta step en modo forward
step(mod_cte,direction="forward",scope=list(lower=mod_cte, upper=mod9))
```

Como resultado, el mtodo forward indica el modelo ptimo contiene las variables independientes $grad_univ$, $paro$, $jubilados$, $asia$, $mormones$, $densidad$, $evangelicos$, $mujeres$, y $protestantes$.

Esto indica que los modelos ptimos no coinciden totalmente, ya que por ejemplo $jubilados$ y $protestantes$ no estn en el modelo de backward y si de forward.

Por tanto, para decidir que modelo es mejor, habr que comparar el AIC final de cada uno. En el caso del primero, el AIC es de 165.22, del segundo es de 166.02.

Entonces, el mdelo ptimo es el conseguido a travs del mtodo $backward$.

7 - Se crea un output, donde se puede dar un vistazo rpido de todos los datos obtenidos, como se consigue con el mtodo $summary$ de R

```r
formula = Girth ~ Height + Volume
variables = all.vars(as.formula(formula))
variables[1]
```

```r
lm_ej5 <- function(data, formula, alpha = 0.05) {
    
    # 0. Declaracin de la funcin y sus parmetros
    # data:  dataframe que contiene los datos
    # formula: frmula del modelo de regresin lineal
    # alpha: nivel de significacin para los intervalos de confianza 
  
    # 1. Extraer los nombres de las variables de la frmula con el metodo all.vars
    variables = all.vars(as.formula(formula))
    variable_dependiente = variables[1]
    variables_independientes = variables[-1]
    
    # 2. Creacin de la matriz X con las variables independientes con B_0s, y el vector y con la variable dependiente
    X = as.matrix(data[, variables[-1]])  # Excluye la variable dependiente
    # Agregar una columna de unos para los B_0
    X = cbind(1, X)
    # Vector y igual a la columna Girth original
    y = as.matrix(data[[variables[1]]])
  
    # 3. Calcular los coeficientes usando la frmula matricial:  = (X^T * X)^(-1) * X^T * y
    coeficientes = solve(t(X) %*% X) %*% (t(X) %*% y)
  
    # 4. Nmero de observaciones y nmero de variables independientes (predictores)
    n = nrow(X)
    p = length(variables) - 1 

    # 5. Extraccion de datos

    # Valores ajustados
    val_ajustado = rep(coeficientes[1], n)
    # Basado en ejemplo boletin 3.3 yajustado <- bhat[1]+bhat[2]*wdata$educ+bhat[3]*wdata$exper+bhat[4]*wdata$tenure
    for (i in seq_along(variables_independientes)) {
        val_ajustado <- val_ajustado + coeficientes[i + 1] * data[[variables_independientes[i]]]
    }
    
    # Grados de libertad
    gdl = n - p - 1
  
    # Residuos
    residuos = y - X %*% coeficientes
  
    # Suma de cuadrados de los residuos
    ssr = sqrt(sum(residuos^2)/gdl)

    # Coeficientes de determinacion, No funciona porque al hacer la media aparece NA por algun motivo
    # residuos2 = data[variable_dependiente] - mean(data[variable_dependiente])
    # coef_det = 1-(sum(residuos^2)/sum((residuos2)^2))
          
    # 6. Crear un objeto de modelo para devolver
    modelo = list(
        coefficients = coeficientes,
        residuals = residuos,
        gdl = gdl,
        ssr = ssr,
        val_ajustado = val_ajustado
        #coef_det = coef_det
    )

    # 7. Output
    cat("Call:", "\n", formula, "\n")
    cat("Residuals:", "\n", "Min:", min(residuos), "1Q:", quantile(residuos, 0.25),
        "Median:", median(residuos), "3Q:", quantile(residuos, 0.75), "Max:", max(residuos), "\n")
    cat("Coefficients:", "\n", coeficientes, "\n")
    cat("Degrees of Freedom:", gdl, "\n")
    cat("Residual Standar Error:", ssr, "\n")
    #cat("Multiple R-squared:", coef_det, "\n")

    
    # Devolver el modelo ajustado
    return(modelo)
}

# Ejemplo de uso:
form = "Girth ~ Height + Volume"  # Ajusta esto segn tus nombres de variables
mod_ej5_func = lm_ej5(data_ej5, form)
cat("Los valores ajustados son:","\n",mod_ej5_func$val_ajustado)
```
